## 4) How many male billionaires are there compared to the number of female billionares? What percent is that? Do they have a different average wealth?

> **TIP:** The last part uses `groupby`, but the count/percent part does not.
> **TIP:** When I say "average," you can pick what kind of average you use.

```python
df['gender'].value_counts()
```

```python
isMale = df['gender'] == 'male'
males = df[isMale]
```

```python
isFemale = df['gender'] == 'female'
females = df[isFemale]
```

```python
maleBillionares = males['networthusbillion'].describe()
maleBillionares
```

```python
femaleBillionares = females['networthusbillion'].describe()
femaleBillionares
```

```python
maleCount = males['networthusbillion'].count()
maleCount
```

```python
maleMean = round(males['networthusbillion'].mean(),2)
maleMean
```

```python
femaleCount = females['networthusbillion'].count()
femaleCount
```

```python
femaleMean = round(females['networthusbillion'].mean(),2)
femaleMean
```

```python
totalCount = maleCount + femaleCount
percentageMale = round((maleCount / totalCount) * 100,2)
percentageFemale = round((femaleCount / totalCount)* 100,2)

if maleCount > femaleCount:
    difference1 = round(maleCount - femaleCount,2)
    print("There are", maleCount, "males, and", femaleCount, "females, so there is", difference1, "more males.")
    print("The percentage of males is", percentageMale, "%",
         "and the percentage of female is", percentageFemale,"%.")
elif maleCount == femaleCount:
    print("There are ", maleCount, "males, and", femaleCount, "females, so there there is the same amount")
    print("The percentage of males is", percentageMale, "%",
         "and the percentage of female is", percentageFemale,"%.")
else:
    difference2 = round(femaleCount - maleCount, 2)
    print("There are", maleCount, "males, and", femaleCount, "females, so there is", difference1, "more females.")
    print("The percentage of males is", percentageMale, "%",
         "and the percentage of female is", percentageFemale,"%.")
```

```python
if maleMean > femaleMean:
    print("Male average wealth is higher at", maleMean, "while females are at", femaleMean)
elif maleMean == femaleMean:
    print("The average wealth is the same for both male and female at", femaleMean)
else:
    print("Female average wealth is higher at", femaleMean, "while males are at", maleMean)
```
